Linear Regression - define dependent and independent variables

```python
y = data['GPA']
x1 = data['SAT']
```

Explore the data

```python
plt.scatter(x1,y)
plt.xlabel("SAT", fontsize = 20)
plt.ylabel("GPA", fontsize = 20)
plt.show()
```

Regression itself

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 0.0017*x1 + 0.2750
fig = plt.plot(x1,yhat, lw = 4, c= "orange", label = "regression")
plt.xlabel("SAT", fontsize = 20)
plt.ylabel("GPA", fontsize = 20)
plt.show()
```

Dummy Variables Example

```python
raw_data = pd.read_csv("Dummies.csv")
```

```python
raw_data
```

```python
ddata = raw_data.copy()
```

```python
ddata['Attendance'] = ddata['Attendance'].map({'Yes':1, 'No':0})
```

```python
ddata
```

```python
ddata.describe()
```

Multiple Linear Regression - New data

```python
yd = ddata['GPA']
xd1 = ddata[['SAT', 'Attendance']]

xd = sm.add_constant(xd1)
xdresults = sm.OLS(yd,xd).fit()
xdresults.summary()
```

Plot the data

```python
plt.scatter(ddata['SAT'],y, c = ddata['Attendance'], cmap ='RdYlGn_r')
yhat_no = 0.6439 + 0.0014*ddata['SAT']
yhat_yes = 0.8556 + 0.0014*ddata['SAT']
fig = plt.plot(ddata['SAT'], yhat_no, lw = 2, c = 'red', label = 'regression line1')
fig = plt.plot(ddata['SAT'], yhat_yes, lw = 2, c = 'blue', label = 'regression line2')
fig = plt.plot(ddata['SAT'], yhat, lw = 2, c= 'green', label = 'regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```
